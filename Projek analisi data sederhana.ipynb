```python
import numpy as np
import pandas as pd
jumlah_data = 1000
Id_Pekerja = np.random.randint(0,1000,size=jumlah_data)
Usia = np.random.randint(20,60,size=jumlah_data)
Jenis_Kelamin = np.random.choice(['L','P'],size=jumlah_data)
Departemen = np.random.choice(['HR','IT','Sales','Finance'],size=jumlah_data)
Pendapatan = np.random.randint(5000000,30000000,size=jumlah_data)
lama_bekerja = np.random.randint(0,30,size=jumlah_data)
Keseimbangan_hidup = np.random.choice(['Poor','Fair','Good','excellent'],size=jumlah_data)
job_level = np.random.choice(['Junior','Mid','Senior'],size=jumlah_data)
Remote_work = np.random.choice(['Ya','Tidak'],size=jumlah_data)
Dapat_Promosi_Kerja =  np.random.choice(['Ya','Tidak'],size=jumlah_data)
Kepuasan_Kerja =  np.random.choice(['Puas','Tidak'],size=jumlah_data)

df = pd.DataFrame ({
    'Id Pekerja' : Id_Pekerja,
    'Usia' : Usia,
    'Jenis Kelammin' : Jenis_Kelamin,
    'Departemen' : Departemen,
    'Pendapatan' : Pendapatan,
    'Lama_Bekerja' : lama_bekerja,
    'Keseimbangan Hidup' : Keseimbangan_hidup,
    'Job Level' : job_level,
    'Remote Work' : Remote_work,
    'Dapaat Promosi' : Dapat_Promosi_Kerja,
    'Kepuasan Kerja' : Kepuasan_Kerja
})

print(df)
print(df.info()) 
print(df.head())

df.to_csv("Data Kepuasan Pekerja.csv")
```
